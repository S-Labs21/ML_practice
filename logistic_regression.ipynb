```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\shakt\ML\social_media-ads\Social_Network_Ads.csv")

data.head()
```

```python
plt.scatter(data.Age, data.EstimatedSalary)
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```

```python
print("Data features: {}".format(data.keys()))
```

```python
features = data[["Age", "EstimatedSalary"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, data.Purchased, test_size=0.3)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
new_data = pd.DataFrame({'Age': [40], 'EstimatedSalary': [33000]})
prediction = model.predict(new_data)
print(f"Prediction for new data: {prediction}")
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay

acc = accuracy_score(y_test, y_pred)
print("Accuracy: {} %".format(acc*100))
```
